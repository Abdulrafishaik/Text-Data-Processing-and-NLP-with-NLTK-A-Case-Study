```python
bussiness=open("/content/business_1.txt","r")
read_data1=bussiness.read()
print(read_data1)
```

```python
# contractions
contract=contractions.fix(read_data1)
print(contract)
```

# **Punctuations**

```python
# punctuation
german_data=[]
for i in contract:
  if i not in string.punctuation:
    german_data.append(i)
    text="".join(german_data)
print(text)
```

```python
# stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# stemming
real_text=[]
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for i in stop_word:
  real_text.append(ps.stem(i))
text=" ".join(real_text)
print(text)
```

# **Stopped words**

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_word=stopwords.words('english')
print(stop_word)
print(len(stop_word))
```

```python
def stop_text(input_text):
  word_text=[]
  text_split=input_text.split()
  for i in text_split:
    if i not in stop_word:
      word_text.append(i)
      text=" ".join(word_text)
  return text

```

```python
word1=stop_text(token_text1)
```

```python
word2=stop_text(token_text2)
word2
```

```python
word3=stop_text(token_text3)
word3
```

```python
word4=stop_text(token_text4)
word4
```

# **Stemming**

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stemmer(input_text):
  stemp_text=[]
  for i in input_text:
    stemp_text.append(ps.stem(i))
    text="".join(stemp_text)
  return text
```

```python
stem1=stemmer(word1)
```

```python
stem2=stemmer(word2)
```

```python
stem3=stemmer(word3)
stem3
```

```python
stem4=stemmer(word4)
stem4
```
